## Q.2 By using sklearn grid search function and find the best parameters of SVM. Use tuning performance metrics as MCC.

```python
from sklearn.metrics import make_scorer, matthews_corrcoef
```

```python

from sklearn.model_selection  import GridSearchCV
```

```python
scorer = make_scorer(matthews_corrcoef)

param_grid = {'C': [1,10,20],
             'gamma': [1,0.1,0.01,0.001],
             'kernel' : ['rbf','linear']}
grid = GridSearchCV(SVC(),param_grid, verbose =1,scoring = scorer)
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_estimator_)
```

```python
grid.predict(x_test)
```

```python
df =pd.DataFrame(grid.cv_results_)
df[['param_C','param_gamma','param_kernel','mean_test_score']]
```

### Q.4 Divide data into 5 fold using NumPy and find mean CV accuracy and mean CV MCC

```python
fold = 5
shuf = np.random.choice(fold,len(x))
for i in range(5):
    X_train = x.iloc[np.where(shuf != i)[0],]
    X_test = x.iloc[np.where(shuf != i)[0],]
    y_train = y.iloc[np.where(shuf != i)[0],]
    y_test = y.iloc[np.where(shuf != i)[0],]
    
    model = SVC(C=1, gamma = 1, kernel = 'linear')
    model.fit(X_train,y_train)
    
    y_pred = model.predict(X_test)
    
    
    print('mcc' ,matthews_corrcoef(y_test,y_pred))
    print("acc",accuracy_score(y_test,y_pred))
    
```

### Q.5 Use sklearn cross val score function to find 5 fold mean CV accuracy and mean CV MCC.

```python
from sklearn.model_selection import cross_val_score
score_acc = cross_val_score(SVC(C=1, gamma = 1, kernel = 'linear'),x,y,cv = 5)
score_acc.mean()
```
